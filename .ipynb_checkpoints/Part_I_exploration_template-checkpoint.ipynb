```python
base_color = sb.color_palette()[0]
prating_counts = df1['ProsperRating (Alpha)'].value_counts()
prating_order = prating_counts.index

sb.countplot(data=df1, x='ProsperRating (Alpha)', color=base_color, order=prating_order);
```

```python
p_list = df1['ProsperRating (Alpha)'].value_counts().sum()

# Considering the same chart from the Example 1 above, print the text (proportion) BELOW the bars
base_color = sb.color_palette()[0]
sb.countplot(data=df1, x='ProsperRating (Alpha)', color=base_color, order=prating_order);


# Recalculating the type_counts just to have clarity.
pr_counts = df1['ProsperRating (Alpha)'].value_counts()

# get the current tick locations and labels
locs, labels = plt.xticks(rotation=0) 

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):

    # get the text property for the label to get the correct count
    count = pr_counts[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/p_list)

    # print the annotation just below the top of the bar
    plt.text(loc, count+2, pct_string, ha = 'center', color = 'black')
```

```python
listing_counts = df1['ListingCategory (numeric)'].value_counts()
listing_order = listing_counts.index

plt.figure(figsize = [14, 6])
ax = sb.countplot(x="ListingCategory (numeric)", data=df1, color=base_color, order=listing_order)

# This function is used to change the count plot numbers into percentages
def without_hue(plot, feature):
    total = len(feature)
    for p in plot.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        plot.annotate(percentage, (x, y), size = 10, va='bottom', ha='center')

without_hue(ax, df1['ListingCategory (numeric)'])
plt.xlabel('listing category');
```

```python
employ_counts = df1['EmploymentStatus'].value_counts()
employ_order = employ_counts.index

plt.figure(figsize = [14, 6])
ax = sb.countplot(data=df1, x="EmploymentStatus", color=base_color, order=employ_order)

without_hue(ax, df1['EmploymentStatus'])
plt.xticks(rotation=0);
```

```python
#LoanStatus
loan_counts = df1['LoanStatus'].value_counts()
loan_order = loan_counts.index

plt.figure(figsize = [14, 6])
ax = sb.countplot(data=df1, x="LoanStatus", color=base_color, order=loan_order)

without_hue(ax, df1['LoanStatus'])
plt.xticks(rotation=45);
```

```python
ticks = df1.IsBorrowerHomeowner.value_counts().index
labels = ['HomeOwner','Not HomeOwner']

ax = sb.countplot(data=df1, x="IsBorrowerHomeowner")
without_hue(ax, df1['IsBorrowerHomeowner'])
plt.xticks(ticks,labels)
plt.xlabel('');
```

```python
# check the relationship between loan outcome status and the monthly loan payment
plt.figure(figsize = [15, 6])
loan_counts = df1['LoanStatus'].value_counts()
loan_order = loan_counts.index

sb.violinplot(data=df1, x='LoanStatus', y='MonthlyLoanPayment', color=base_color, inner='quartile',order=loan_order)
plt.xticks(rotation=15);
```

```python
# check the relationship between loan outcome status and Home Owner
plt.figure(figsize = [15, 6])
legend_labels = ['Not HomeOwner', 'Is HomeOwner']

ax = sb.countplot(data = df1, x = 'LoanStatus', hue = 'IsBorrowerHomeowner', order=loan_order)
without_hue(ax, df1['LoanStatus'])
# loc = 'upper left'
plt.legend(labels = legend_labels, title = 'Home Ownership Status')
plt.xticks(rotation=15);
```

```python
# check the relationship between loan outcome status and Borrower Rate
plt.figure(figsize = [15, 6])
sb.violinplot(data=df1, x='LoanStatus', y='BorrowerRate', color=base_color, inner='quartile',order=loan_order)
plt.xticks(rotation=15);
```

```python
# check the relationship between loan outcome status and Employment status

plt.figure(figsize = [15, 6])
g = sb.FacetGrid(data = df1, col = 'LoanStatus', col_wrap = 4, size = 4)
g.map(sb.countplot, 'EmploymentStatus', order=employ_order)
plt.xticks(rotation=15);
```

```python
# check the relationship between Borrower rate and StatedMonthlyIncome
plt.figure(figsize = [20, 5])
sample1 = df1.sample(n = 15000)
sample1.shape

plt.subplot(1, 2, 1)
sb.regplot(data = sample1, x = 'StatedMonthlyIncome', y = 'BorrowerRate', truncate=False, x_jitter=0.3,scatter_kws={'alpha':1/10});

plt.subplot(1, 2, 2)
plt.hist2d(data = df1, x = 'StatedMonthlyIncome', y = 'BorrowerRate')
plt.colorbar();
```
